This creates a csv with information about the stations and the pollutants that measured during 2015.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../datos/loc_estaciones/estaciones.csv", index_col = 0)
```

```python
contaminantes = ["O3","NO2","NO","NOx","CO","SO2","PM10","PM2.5"]
```

```python
for poll in contaminantes:
    df[poll] = np.ones(43)
```

```python
cont = pd.read_csv('../datos/contaminantes/2015/contaminantes_2015.CSV', skiprows = 10)
```

```python
for station in df.index:
    print('--', station)
    for poll in contaminantes:
        aux_i = cont.loc[cont['id_station'] == station].loc[cont['id_parameter'] == poll].shape[0] != 0
        #if aux_i == True:
        #    df.at[station, poll] = int(1)
        if aux_i == False:
            df.at[station, poll] = np.nan
            
```
